```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("hotel_bookings.csv", header=0)
```

```python
data.columns
```

```python
hotel_name = data[["hotel", "is_canceled"]]
print(hotel_name[:10])
```

```python
data.describe()
```

```python
agent = data["agent"]
agent[:10]
```

```python
data.hist(column="agent", grid=False, figsize=(20, 10), bins=100)
```

```python
agent.dropna(inplace=True)
nagents = int(max(agent))
# nagents = int(max(agent.dropna()))
data.hist(column="agent", grid=False, figsize=(20, 10), bins=nagents)
```

```python
count_by_agent = data.groupby("agent").count()
fig = plt.figure(figsize=(20, 10))
plt.bar(x=count_by_agent.index, height=count_by_agent["hotel"])
plt.show()
```

```python
count_by_agent.head()
```

```python
fig2 = plt.figure(figsize=(20, 10))
plt.boxplot(count_by_agent["hotel"])
plt.show()
```

```python
count_by_agent_filtered = count_by_agent[count_by_agent["hotel"] <= 200]
fig2 = plt.figure(figsize=(20, 10))
plt.boxplot(count_by_agent_filtered["hotel"])
plt.show()
```
